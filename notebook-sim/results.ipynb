### ``Multi-Agent Simulation``

``C++``

```python
import pandas as pd
import matplotlib.pyplot as plt

fichiers_csv = [f'../cpp/simulation/simulation_data{i}.csv' for i in range(1, 101)]

dfs = []

for fichier in fichiers_csv:
    df_current = pd.read_csv(fichier)
    dfs.append(df_current)

df_mean = pd.concat(dfs).groupby('day').mean().reset_index()

print(df_mean.columns)
plt.figure(figsize=(10, 6))

plt.plot(df_mean['day'], df_mean['susceptibleCount'], label="Susceptible")
plt.plot(df_mean['day'], df_mean['exposedCount'], label="Exposed")
plt.plot(df_mean['day'], df_mean['infectedCount'], label="Infected")
plt.plot(df_mean['day'], df_mean['recoveredCount'], label="Recovered")

plt.title("(C++) Average simulation of a disease spread")
plt.xlabel("Day")
plt.ylabel("Count (average)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
overall_mean = df_mean[['susceptibleCount', 'exposedCount', 'infectedCount', 'recoveredCount']].mean(axis=0)
overall_std = df_mean[['susceptibleCount', 'exposedCount', 'infectedCount', 'recoveredCount']].std(axis=0)
overall_variance = df_mean[['susceptibleCount', 'exposedCount', 'infectedCount', 'recoveredCount']].var(axis=0)

statistics_df = pd.DataFrame({
    'Category': ['Susceptible', 'Exposed', 'Infected', 'Recovered'],
    'Mean': overall_mean.values,
    'Standard deviation': overall_std.values,
    'Variance': overall_variance.values
})

statistics_df.set_index('Category', inplace=True)
statistics_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dictionnaire pour stocker les DataFrame de chaque langage
data_par_langage = {
    'Ruby': [f'../ruby/src/simulation/simulation_data{i}.csv' for i in range(1, 101)],
    'Rust': [f'../rust/simulation/simulation_data{i}.csv' for i in range(1, 101)],
    'Java': [f'../java/src/simulation/simulation_data{i}.csv' for i in range(1, 101)],
    'C++': [f'../cpp/simulation/simulation_data{i}.csv' for i in range(1, 101)],
}

# Initialisation de la figure
plt.figure(figsize=(10, 10))

# Couleurs pour chaque langage
couleurs = {
    'Ruby': 'red',
    'Rust': 'blue',
    'Java': 'green',
    'C++': 'orange',
}

# Traitement pour chaque langage
for langage, fichiers in data_par_langage.items():
    dfs = [pd.read_csv(fichier) for fichier in fichiers]
    df_mean = pd.concat(dfs).groupby('day').mean().reset_index()
    
    plt.plot(df_mean['day'], df_mean['susceptibleCount'], label=f"S ({langage})", color=couleurs[langage])
    plt.plot(df_mean['day'], df_mean['exposedCount'], label=f"E ({langage})", color=couleurs[langage], linestyle='--')
    plt.plot(df_mean['day'], df_mean['infectedCount'], label=f"I ({langage})", color=couleurs[langage], linestyle='-.')
    plt.plot(df_mean['day'], df_mean['recoveredCount'], label=f"R ({langage})\n", color=couleurs[langage], linestyle=':')
    plt.legend(loc='upper right', bbox_to_anchor=(1, 1), borderaxespad=0.)
    plt.tight_layout()
    
plt.title("Simulation of disease spread by programming language")
plt.xlabel("Day")
plt.ylabel("Count (average)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Execution time
langages = ['C++', 'Java', 'Python', 'Ruby', 'Rust']
real = [7 + 11.738/60, 11 + 9.814/60, 82 + 23.372/60, 72 + 0.688/60, 23 + 35.738/60]
user = [7 + 8.532/60, 23 + 22.135/60, 82 + 22.150/60, 71 + 52.359/60, 23 + 31.640/60]
sys = [3.191/60, 10.305/60, 1.150/60, 7.941/60, 3.840/60]
fig, axs = plt.subplots(3, 1, figsize=(14, 12), sharex=True)
colors = ['skyblue', 'lightgreen', 'salmon']

# Real
axs[0].bar(langages, real, color=colors[0])
axs[0].set_title('Real time')
axs[0].set_ylabel('Minutes')

# User
axs[1].bar(langages, user, color=colors[1])
axs[1].set_title('User time')
axs[1].set_ylabel('Minutes')

# Sys
axs[2].bar(langages, sys, color=colors[2])
axs[2].set_title('System. time')
axs[2].set_ylabel('Minutes')

for ax in axs:
    ax.set_xlabel('Programming languages')
    ax.grid(True, linestyle='--', which='major', color='grey', alpha=0.5)
    
plt.tight_layout()
plt.xticks(range(len(langages)), langages)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
languages = ['C++', 'Java', 'Python', 'Ruby', 'Rust']
average_time = [607.82, 651.97, 5471.22, 3698.00, 1285.72]
average_cpu_time = [607.44, 1422.78, 5470.40, 3697.55, 1285.48]
average_memory_usage = [549.26171875, 2091.8046875, 1731.19140625, 2401.6875, 546.75]

# Creating the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Setting positions for the bars
positions = np.arange(len(languages))
bar_width = 0.25  # width of the bars

# Plotting the average time and average CPU time bars side by side
average_time_bars = ax.bar(positions - bar_width / 2, average_time, bar_width, label='Average Time (sec)', color='skyblue')
average_cpu_time_bars = ax.bar(positions + bar_width / 2, average_cpu_time, bar_width, label='Average CPU Time (sec)', color='lightgreen')

# Adding labels and title
ax.set_xlabel('Programming Language')
ax.set_ylabel('Seconds')
ax.set_title('Average Execution Time vs. Average CPU Time')
ax.set_xticks(positions)
ax.set_xticklabels(languages)

# Twinning the y-axis to plot the average memory usage
ax2 = ax.twinx()
average_memory_usage_line = ax2.plot(languages, average_memory_usage, color='r', marker='o', label='Average Memory Usage (KB)')
ax2.set_ylabel('Memory Usage (KB)', color='r')

# Creating a custom legend
legend_elements = [plt.Line2D([0], [0], color='skyblue', lw=4, label='Average Time (sec)'),
                   plt.Line2D([0], [0], color='lightgreen', lw=4, label='Average CPU Time (sec)'),
                   plt.Line2D([0], [0], color='r', marker='o', label='Average Memory Usage (KB)')]
ax.legend(handles=legend_elements, loc='upper left')

plt.show()
```
