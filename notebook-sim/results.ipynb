```python
import pandas as pd
import matplotlib.pyplot as plt

fichiers_csv = [f'../cpp/simulation/simulation_data{i}.csv' for i in range(1, 101)]

dfs = []

for fichier in fichiers_csv:
    df_current = pd.read_csv(fichier)
    dfs.append(df_current)

df_mean = pd.concat(dfs).groupby('day').mean().reset_index()

print(df_mean.columns)
plt.figure(figsize=(10, 6))

plt.plot(df_mean['day'], df_mean['susceptibleCount'], label="Susceptible")
plt.plot(df_mean['day'], df_mean['exposedCount'], label="Exposed")
plt.plot(df_mean['day'], df_mean['infectedCount'], label="Infected")
plt.plot(df_mean['day'], df_mean['recoveredCount'], label="Recovered")

plt.title("(C++) Average simulation of a disease spread")
plt.xlabel("Day")
plt.ylabel("Count (average)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
overall_mean = df_mean[['susceptibleCount', 'exposedCount', 'infectedCount', 'recoveredCount']].mean(axis=0)
overall_std = df_mean[['susceptibleCount', 'exposedCount', 'infectedCount', 'recoveredCount']].std(axis=0)
overall_variance = df_mean[['susceptibleCount', 'exposedCount', 'infectedCount', 'recoveredCount']].var(axis=0)

statistics_df = pd.DataFrame({
    'Category': ['Susceptible', 'Exposed', 'Infected', 'Recovered'],
    'Mean': overall_mean.values,
    'Standard deviation': overall_std.values,
    'Variance': overall_variance.values
})

statistics_df.set_index('Category', inplace=True)
statistics_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dictionnaire pour stocker les DataFrame de chaque langage
data_par_langage = {
    'Ruby': [f'../ruby/src/simulation/simulation_data{i}.csv' for i in range(1, 101)],
    'Rust': [f'../rust/simulation/simulation_data{i}.csv' for i in range(1, 101)],
    'Python': [f'../python/simulation/simulation_data{i}.csv' for i in range(1, 101)],
    'Java': [f'../java/src/simulation/simulation_data{i}.csv' for i in range(1, 101)],
    'C++': [f'../cpp/simulation/simulation_data{i}.csv' for i in range(1, 101)],
}

# Initialisation de la figure
plt.figure(figsize=(10, 10))

# Couleurs pour chaque langage
couleurs = {
    'Ruby': 'red',
    'Rust': 'blue',
    'Python': 'purple',
    'Java': 'green',
    'C++': 'orange',
}

# Traitement pour chaque langage
for langage, fichiers in data_par_langage.items():
    dfs = [pd.read_csv(fichier) for fichier in fichiers]
    df_mean = pd.concat(dfs).groupby('day').mean().reset_index()
    
    plt.plot(df_mean['day'], df_mean['susceptibleCount'], label=f"S ({langage})", color=couleurs[langage])
    plt.plot(df_mean['day'], df_mean['exposedCount'], label=f"E ({langage})", color=couleurs[langage], linestyle='--')
    plt.plot(df_mean['day'], df_mean['infectedCount'], label=f"I ({langage})", color=couleurs[langage], linestyle='-.')
    plt.plot(df_mean['day'], df_mean['recoveredCount'], label=f"R ({langage})", color=couleurs[langage], linestyle=':')
    plt.legend(loc='upper right', bbox_to_anchor=(0, 1), borderaxespad=0.)
    plt.tight_layout()
    
plt.title("Simulation of disease spread by programming language")
plt.xlabel("Day")
plt.ylabel("Count (average)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def calculate_mean_for_language(files_pattern, num_files):
    dfs = []
    for i in range(1, num_files + 1):
        file_path = files_pattern.format(i)
        df = pd.read_csv(file_path)
        dfs.append(df)
    combined_df = pd.concat(dfs)
    mean_df = combined_df.groupby('day').mean().reset_index()
    return mean_df


mean_cpp = calculate_mean_for_language("../cpp/simulation/simulation_data{}.csv", 100)
mean_java = calculate_mean_for_language("../java/src/simulation/simulation_data{}.csv", 100)
mean_python = calculate_mean_for_language("../python/simulation/simulation_data{}.csv", 100)
mean_ruby = calculate_mean_for_language("../ruby/src/simulation/simulation_data{}.csv", 100)
mean_rust = calculate_mean_for_language("../rust/simulation/simulation_data{}.csv", 100)


df_cpp_vs_java = pd.DataFrame({
        'S': [],
        'E': [],
        'I': [],
        'R' :[]
    })

df_cpp_vs_python = pd.DataFrame({
        'S': [],
        'E': [],
        'I': [],
        'R' :[]
    })

df_cpp_vs_ruby = pd.DataFrame({
        'S': [],
        'E': [],
        'I': [],
        'R' :[]
    })

df_cpp_vs_rust = pd.DataFrame({
        'S': [],
        'E': [],
        'I': [],
        'R' :[]
    })

df_cpp_vs_java['S'] = abs(mean_cpp['susceptibleCount'] - mean_java['susceptibleCount'])
df_cpp_vs_java['E'] = abs(mean_cpp['exposedCount'] - mean_java['exposedCount'])
df_cpp_vs_java['I'] = abs(mean_cpp['infectedCount'] - mean_java['infectedCount'])
df_cpp_vs_java['R'] = abs(mean_cpp['recoveredCount'] - mean_java['recoveredCount'])

df_cpp_vs_python['S'] = abs(mean_cpp['susceptibleCount'] - mean_python['susceptibleCount'])
df_cpp_vs_python['E'] = abs(mean_cpp['exposedCount'] - mean_python['exposedCount'])
df_cpp_vs_python['I'] = abs(mean_cpp['infectedCount'] - mean_python['infectedCount'])
df_cpp_vs_python['R'] = abs(mean_cpp['recoveredCount'] - mean_python['recoveredCount'])

df_cpp_vs_ruby['S'] = abs(mean_cpp['susceptibleCount'] - mean_ruby['susceptibleCount'])
df_cpp_vs_ruby['E'] = abs(mean_cpp['exposedCount'] - mean_ruby['exposedCount'])
df_cpp_vs_ruby['I'] = abs(mean_cpp['infectedCount'] - mean_ruby['infectedCount'])
df_cpp_vs_ruby['R'] = abs(mean_cpp['recoveredCount'] - mean_ruby['recoveredCount'])

df_cpp_vs_rust['S'] = abs(mean_cpp['susceptibleCount'] - mean_rust['susceptibleCount'])
df_cpp_vs_rust['E'] = abs(mean_cpp['exposedCount'] - mean_rust['exposedCount'])
df_cpp_vs_rust['I'] = abs(mean_cpp['infectedCount'] - mean_rust['infectedCount'])
df_cpp_vs_rust['R'] = abs(mean_cpp['recoveredCount'] - mean_rust['recoveredCount'])

for column in df_cpp_vs_java:
    plt.plot(df_cpp_vs_java.index, df_cpp_vs_java[column], label=column)

plt.title('CPP vs Java')
plt.xlabel('Days')
plt.ylabel('Differences')
plt.legend()
plt.show()

df_cpp_vs_java
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

data = np.vstack((real, user, sys)).T

kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
clusters = kmeans.labels_

scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_normalized)

fig, ax = plt.subplots(figsize=(10, 7))
scatter = ax.scatter(data_pca[:, 0], data_pca[:, 1], c=clusters, cmap='viridis', s=100, alpha=0.6, edgecolors='w', linewidths=2)

for i, txt in enumerate(langages):
    ax.text(data_pca[i, 0], data_pca[i, 1], txt, color='black')

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('PCA based on performance (Memory Usage, Real Time & CPU Time)')
plt.grid(True, linestyle='--', color='grey', alpha=0.5)
plt.show()
```
